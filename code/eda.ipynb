## 3. Exploratory data analysis

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
```

```python
pd.set_option('display.max_columns', 500)
```

```python
warnings.filterwarnings('ignore', category = RuntimeWarning)
warnings.filterwarnings('ignore', category = UserWarning)
```

```python
sns.set_theme(style = 'darkgrid')
```

```python
# load training data

X_train = pd.read_pickle('X_train.pkl')
y_train = pd.read_pickle('y_train.pkl')
```

```python
full_train = X_train.copy()
full_train['malig_infarct'] = y_train
```

```python
cont_vars = []
cat_vars = []

for var in X_train.columns:
    if X_train[var].dtype == 'float64' or X_train[var].dtype == 'int64':
        cont_vars.append(var)
    elif X_train[var].dtype == 'O':
        cat_vars.append(var)
    else:
        print('{} is not either variable type'.format(var))
```

### Univariate analysis of features - distributions

```python
X_train[cont_vars].describe()
```

```python
for var in cont_vars:
    plt.figure()
    sns.displot(X_train[var], kde = True)
```

```python
X_train[cat_vars].describe()
```

```python
for var in cat_vars:
    plt.figure()
    sns.countplot(x = var, data = X_train)
```

### Bivariate analysis among features

```python
g = sns.PairGrid(X_train[cont_vars])
g.map(sns.scatterplot)
g.map_offdiag(sns.scatterplot)
```

```python
# plot a scatterplot heatmap

corr_mat = X_train[cont_vars].corr().stack().reset_index(name = 'correlation')

g = sns.relplot(
    data = corr_mat,
    x = 'level_0', y = 'level_1', hue = 'correlation', size = 'correlation',
    palette = 'vlag', hue_norm = (-1, 1), edgecolor = '.7',
    height = 10, sizes = (50, 250), size_norm = (-.2, .8),
)

g.set(xlabel = '', ylabel = '', aspect = 'equal')
g.despine(left = True, bottom = True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
for artist in g.legend.legendHandles:
    artist.set_edgecolor('.7')
```

```python
# plot the individual scatterplots for variables with r > 0.6

for rnum in range(corr_mat.shape[0]):
    if corr_mat.loc[rnum, 'correlation'] > 0.6 and not corr_mat.loc[rnum, 'correlation'] == 1:
        plt.figure()
        sns.relplot(x = corr_mat.loc[rnum, 'level_0'], y = corr_mat.loc[rnum, 'level_1'], data = X_train, 
                   alpha = 0.5).set(title = 'correlation = {0:.2f}'.format(corr_mat.loc[rnum, 'correlation']))
```

```python
def plot_bar_percent(x, y, data):
    plt.figure()
    g = data.groupby(x)[y].value_counts(normalize=True).mul(100).rename('percent').reset_index().pipe((
        sns.catplot, 'data'), x = x, y = 'percent', hue = y, kind = 'bar')
    g.ax.set_ylim(0, 100)
    
    # insert percentages above bars 
    for p in g.ax.patches:
        txt = str(p.get_height().round(2)) + '%'
        txt_x = p.get_x() 
        txt_y = p.get_height()
        g.ax.text(txt_x,txt_y,txt)
```

```python
for var in cat_vars:
    x = 'malig_infarct'
    y = var
    plot_bar_percent(x = x, y = y, data = full_train)
```

### Decrease levels of certain features and re-evaluate relationship with target

```python
plot_bar_percent(x = 'malig_infarct', y = 'tici_success', data = full_train)
```

```python
# dichotomize collateral score

X_train['coll_full'] = np.where(X_train['coll_score'] == 3.0, 'Y', 'N')
full_train['coll_full'] = np.where(full_train['coll_score'] == 3.0, 'Y', 'N')
```

```python
plot_bar_percent(x = 'malig_infarct', y = 'coll_full', data = full_train)
```

```python
# dichotomize pre-morbid mRS
# mRS = 0 compared to 1-5

X_train['pre_mrs_0'] = np.where(X_train['pre_mrs'] == 0.0, 'Y', 'N')
full_train['pre_mrs_0'] = np.where(full_train['pre_mrs'] == 0.0, 'Y', 'N')
```

```python
plot_bar_percent(x = 'malig_infarct', y = 'pre_mrs_0', data = full_train)
```

```python
# dichotomize pre-morbid mRS
# mRS = 0-2 compared to 3-5

X_train['pre_mrs_02'] = np.where(X_train['pre_mrs'] <= 2.0, 'Y', 'N')
full_train['pre_mrs_02'] = np.where(full_train['pre_mrs'] <= 2.0, 'Y', 'N')
```

```python
plot_bar_percent(x = 'malig_infarct', y = 'pre_mrs_02', data = full_train)
```

### Multivariate analysis with target

```python
def plot_multivar_facet(pivot_var, col_order, cont_vars, cat_vars, target):
    print('Pivoting by: {}'.format(pivot_var))
    
    for var in cont_vars:
        plt.figure()
        g = sns.FacetGrid(full_train, col = pivot_var, col_order = col_order)
        g.map(sns.boxplot, target, var)
    
    for var in cat_vars:
        plt.figure()
        sns.catplot(x = target, hue = var, col = pivot_var, data = full_train, kind = 'count', 
                    col_order = col_order)
```

Outcomes by TICI reperfusion

```python
plot_multivar_facet(pivot_var = 'tici', 
                   col_order = ['3', '2B', '2A', '1', '0'], 
                   cont_vars = ['nih_admit', 'cbf_30', 'mismatch_vol'], 
                   cat_vars = ['tpa', 'coll_score'], 
                   target = 'malig_infarct')
```

```python
# dichotomized TICI

plot_multivar_facet(pivot_var = 'tici_success', 
                   col_order = ['Y', 'N'], 
                   cont_vars = ['nih_admit', 'cbf_30', 'mismatch_vol'], 
                   cat_vars = ['tpa', 'coll_score'], 
                   target = 'malig_infarct')
```
